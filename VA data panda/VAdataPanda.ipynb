```python
#main imports
import numpy as np
import pandas as pd
import json
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'browser'#will open in a chrome browser
#3 main files:
#the map of VA
#the amount of infections
#the reopening plan

```

```python
VAmap = json.load(open("files/countiesVA.json", 'r'))
#VAcounties['features'][1] is an example of getting features of Richmond City
```

```python
nameID = {} # create a dictionary to store the id
for feature in VAmap['features']:
    feature['id'] = feature['properties']['NAME']
    nameID[feature['properties']['COUNTYFP']] = feature['id']
    #county FP is the FIPS code for the counties
nameID
```

```python
infections = pd.read_csv('files/infdata/covid_8-7.csv')
infections["Number of Cases"] = (infections["Number of Cases"] / 1000) #divide all cases by 1000

#data 
VAplan = pd.read_csv('files/VAplan2.csv')
VAplan = VAplan[pd.notnull(VAplan['Risk'])]
VAplan['locations'] = VAplan['locations'].apply(lambda x: (x.split("/")[0].replace(",","")))
VAmap['features'][4]['properties'] #countyFP always changes
```

```python
#map create
fig = px.choropleth(
    VAplan,
    locations="locations",
    geojson="files/countiesVA.json",
    color="Risk",
    hover_data=["Risk"],
    title="Estimated Risk per Virginia County for Opening Schools"
)

fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
